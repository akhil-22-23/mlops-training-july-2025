```python
# Initilaizing libraries
import pandas as pd
```

```python
# Reading data directly from source without downloading it to your local system
file = r'https://raw.githubusercontent.com/edquestofficial/mlops-training/refs/heads/main/02_day2/data/train.csv'
house_pricing_data_df = pd.read_csv(file)

# Change the output window size with 'eval_js' library
from google.colab.output import eval_js
eval_js('google.colab.output.setIframeHeight("300")')

# Display relevant information about dataset
house_pricing_data_df.info()
```

```python
# Display percentage of null values per column
eval_js('google.colab.output.setIframeHeight("300")')
(round(house_pricing_data_df.isna().sum()/house_pricing_data_df.shape[0],4)*100).sort_values(ascending=False)
```

```python
## NULL VALUE TREATMENT
# Drop columns that have over 30% null values
eval_js('google.colab.output.setIframeHeight("300")')
columns_isna_percent_df = (round(house_pricing_data_df.isna().sum()/house_pricing_data_df.shape[0],4)*100).sort_values(ascending=False).reset_index().rename(columns={'index':'columns',0:'percent'})
columns_isna_percent_over30_df = columns_isna_percent_df[columns_isna_percent_df['percent']>=30]
columns_isna_percent_over30_df

house_pricing_data_nvt_df = house_pricing_data_df.copy().drop(columns=columns_isna_percent_over30_df['columns'])
house_pricing_data_nvt_df.info()
```

```python
# EDA of dataframe columns
coltype_object = house_pricing_data_nvt_df.select_dtypes(include='object').columns.tolist()
coltype_numeric = house_pricing_data_nvt_df.select_dtypes(include='number').columns.tolist()
print(f'Object type columns: {coltype_object}')
print(f'Numeric type columns: {coltype_numeric}')
```
